#Hierarchical

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(data.iloc[:,1:])
```

```python
df_norm.iloc[:,:]
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
plt.figure(figsize=(10,7))
```

#Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
data2 =pd.read_csv("/content/crime_data.csv")
data2
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(data2.iloc[:,1:])
df_norm
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data2['Clust']=x
```

```python
data2
```

```python
data2.iloc[:,1:5].groupby(data2.Clust).mean()
```

#DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data3 =pd.read_csv("/content/crime_data.csv")
data3
```

```python
data3.info()
```

```python
df=data3.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([data3,cl],axis=1)
```
